```python
#Clean confidence intervals
import re
import unidecode
def clean_value(value):
    value = unidecode.unidecode(value).replace(' ','')
    if value == '0':
        return 0
    else:
        match = re.search(r'(\d+\s*)+(?=\[)', value)
        if match:
            return int((''.join(match.group())))
        else:
            return None

# Apply the custom function to the column
mal_deaths['clean_deaths'] = mal_deaths['raw_deaths'].apply(clean_value)
mal_deaths.head()
```

```python
#load mal cases
mal_cases = pd.read_csv('/content/drive/MyDrive/Data Wrangling Data/malaria_cases.csv')
mal_cases.rename(columns={'First Tooltip': 'raw_cases'}, inplace=True)
mal_cases.head()
```

```python
#clean confidence intervals in mal cases
mal_cases['clean_cases'] = mal_cases['raw_cases'].apply(clean_value)
mal_cases.head()
```
